```python
import warnings
import pandas as pd
import hvplot.pandas
import hvplot as hv
```

```python
import bokeh
bokeh.sampledata.download()
from bokeh.sampledata import stocks
print('All sample data was downloaded')
```

```python
type(stocks)
```

```python
df = pd.DataFrame(getattr(stocks, 'AAPL'))
df['date'] = pd.to_datetime(df.date)
```

```python
df
```

```python
# warning
df.hvplot(x='date', y=['low', 'high'], width=500, subplots=True).cols(2)
```

```python
df.hvplot(x='date', y=['low'], width=500, subplots=True) + df.hvplot(x='date', y=['high'], width=500, subplots=True)
```

```python
df.hvplot(x='date', y=['low'], width=1000, subplots=True) * df.hvplot(x='date', y=['high'], width=1000, subplots=True)
```
